```python
import numpy as np
import os
import csv
import configparser
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator, NullFormatter)


# PbPb
sim_runs = ({'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_0/',  'title': 'MB trig 100ns'},
            #{'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_1/',  'title': 'MB trig 100ns - 100 kHz'},
            #{'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_2/',  'title': 'MB trig 100ns - 150 kHz'},
            #{'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_3/',  'title': 'MB trig 100ns - 200 kHz'},
            #{'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_4/',  'title': 'Periodic 5us/trig 100ns - 50 kHz'},
            #{'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_5/',  'title': 'Periodic 5us/trig 100ns - 100 kHz'},
            #{'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_6/',  'title': 'Periodic 5us/trig 100ns - 150 kHz'},
            #{'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_7/',  'title': 'Periodic 5us/trig 100ns - 200 kHz'},
            {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_8/',  'title': 'Periodic trig 5us'},
            #{'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_9/',  'title': 'Periodic trig 5us - 100 kHz'},
            #{'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_10/', 'title': 'Periodic trig 5us - 150 kHz'},
            #{'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_11/', 'title': 'Periodic trig 5us - 200 kHz'},
            {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_12/', 'title': 'Periodic trig 10us'},
            #{'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_13/', 'title': 'Periodic trig 10us - 100 kHz'},
            #{'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_14/', 'title': 'Periodic trig 10us - 150 kHz'},
            #{'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_15/', 'title': 'Periodic trig 10us - 200 kHz'},
            {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_16/', 'title': 'Periodic trig 20us'},
            #{'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_17/', 'title': 'Periodic trig 20us - 100 kHz'},
            #{'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_18/', 'title': 'Periodic trig 20us - 150 kHz'},
            #{'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_19/', 'title': 'Periodic trig 20us - 200 kHz'},
            {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_20/', 'title': 'Periodic cont 5us'},
            #{'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_21/', 'title': 'Periodic cont 5us - 100 kHz'},
            #{'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_22/', 'title': 'Periodic cont 5us - 150 kHz'},
            #{'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_23/', 'title': 'Periodic cont 5us - 200 kHz'},
            {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_24/', 'title': 'Periodic cont 10us'},
            #{'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_25/', 'title': 'Periodic cont 10us - 100 kHz'},
            #{'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_26/', 'title': 'Periodic cont 10us - 150 kHz'},
            #{'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_27/', 'title': 'Periodic cont 10us - 200 kHz'},
            {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_28/', 'title': 'Periodic cont 20us'}
            #{'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_29/', 'title': 'Periodic cont 20us - 100 kHz'},
            #{'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_30/', 'title': 'Periodic cont 20us - 150 kHz'},
            #{'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_31/', 'title': 'Periodic cont 20us - 200 kHz'}
            )

# PbPb 50 kHz
# sim_runs = ({'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_0/',  'title': 'MB trig 100ns'},
#             {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_8/',  'title': 'Periodic trig 5us'},
#             {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_20/', 'title': 'Periodic cont 5us'},
#             {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_12/', 'title': 'Periodic trig 10us'},
#             {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_24/', 'title': 'Periodic cont 10us'},
#             {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_16/', 'title': 'Periodic trig 20us'},
#             {'event_rate': 50,  'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_28/', 'title': 'Periodic cont 20us'}
#             )
# sim_type = "PbPb"
# sim_additional_title = "50 kHz Pb-Pb"
# plot_output_dir = '/scratch/ITS_SystemC/plots_PbPb/data_rate_plots/'


# PbPb 100 kHz
#sim_runs = ({'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_1/',  'title': 'MB trig 100ns'},
#            {'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_9/',  'title': 'Periodic trig 5us'},
#            {'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_21/', 'title': 'Periodic cont 5us'},
#            {'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_13/', 'title': 'Periodic trig 10us'},
#            {'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_25/', 'title': 'Periodic cont 10us'},
#            {'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_17/', 'title': 'Periodic trig 20us'},
#            {'event_rate': 100, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_29/', 'title': 'Periodic cont 20us'}
#            )
#sim_type = "PbPb"
#sim_additional_title = "100 kHz Pb-Pb"
#plot_output_dir = '/scratch/ITS_SystemC/plots_PbPb/data_rate_plots/'


#sim_runs = ({'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_2/',  'title': 'MB trig 100ns'},
#            {'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_10/', 'title': 'Periodic trig 5us'},
#            {'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_22/', 'title': 'Periodic cont 5us'},
#            {'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_14/', 'title': 'Periodic trig 10us'},
#            {'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_26/', 'title': 'Periodic cont 10us'},
#            {'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_18/', 'title': 'Periodic trig 20us'},
#            {'event_rate': 150, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_30/', 'title': 'Periodic cont 20us'}
#            )
#sim_type = "PbPb"
#sim_additional_title = "150 kHz Pb-Pb"
#plot_output_dir = '/scratch/ITS_SystemC/plots_PbPb/data_rate_plots/'


sim_runs = (
            {'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_3/',  'title': 'MB trig 100ns'},
            {'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_11/', 'title': 'Periodic trig 5us'},
            {'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_23/', 'title': 'Periodic cont 5us'},
            {'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_15/', 'title': 'Periodic trig 10us'},
            {'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_27/', 'title': 'Periodic cont 10us'},
            {'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_19/', 'title': 'Periodic trig 20us'},
            {'event_rate': 200, 'path': '/scratch/ITS_SystemC/sim_output_its_PbPb/run_31/', 'title': 'Periodic cont 20us'}
            )
sim_type = "PbPb"
sim_additional_title = "200 kHz Pb-Pb"
plot_output_dir = '/scratch/ITS_SystemC/plots_PbPb/data_rate_plots/'

# proton-proton
# sim_runs = (
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_0/',  'title': 'MB trig 100ns - 400 kHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_1/',  'title': 'MB trig 100ns - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_2/',  'title': 'MB trig 100ns - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_3/',  'title': 'MB trig 100ns - 5 MHz'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_4/',  'title': 'Periodic 5us/trig 100ns - 400 kHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_5/',  'title': 'Periodic 5us/trig 100ns - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_6/',  'title': 'Periodic 5us/trig 100ns - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_7/',  'title': 'Periodic 5us/trig 100ns - 5 MHz'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_8/',  'title': 'Periodic trig 5us - 400 kHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_9/',  'title': 'Periodic trig 5us - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_10/', 'title': 'Periodic trig 5us - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_11/', 'title': 'Periodic trig 5us - 5 MHz'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_12/', 'title': 'Periodic trig 10us - 400 kHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_13/', 'title': 'Periodic trig 10us - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_14/', 'title': 'Periodic trig 10us - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_15/', 'title': 'Periodic trig 10us - 5 MHz'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_16/', 'title': 'Periodic trig 20us - 400 kHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_17/', 'title': 'Periodic trig 20us - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_18/', 'title': 'Periodic trig 20us - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_19/', 'title': 'Periodic trig 20us - 5 MHz'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_20/', 'title': 'Periodic cont 5us - 400 kHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_21/', 'title': 'Periodic cont 5us - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_22/', 'title': 'Periodic cont 5us - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_23/', 'title': 'Periodic cont 5us - 5 MHz'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_24/', 'title': 'Periodic cont 10us - 400 kHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_25/', 'title': 'Periodic cont 10us - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_26/', 'title': 'Periodic cont 10us - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_27/', 'title': 'Periodic cont 10us - 5 MHz'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_28/', 'title': 'Periodic cont 20us - 400 kHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_29/', 'title': 'Periodic cont 20us - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_30/', 'title': 'Periodic cont 20us - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_31/', 'title': 'Periodic cont 20us - 5 MHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_32/', 'title': 'Periodic cont 1us - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_33/', 'title': 'Periodic cont 1us - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_34/', 'title': 'Periodic cont 1us - 5 MHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_35/', 'title': 'Periodic trig 1us - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_36/', 'title': 'Periodic trig 1us - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_37/', 'title': 'Periodic trig 1us - 5 MHz'},
#     {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_38/', 'title': 'Periodic 1us/trig 100ns - 1 MHz'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_39/', 'title': 'Periodic 1us/trig 100ns - 2 MHz'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_40/', 'title': 'Periodic 1us/trig 100ns - 5 MHz'}
# )

# 5 MHz pp
# sim_runs = (
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_3/',  'title': 'MB trig 100ns'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_37/', 'title': 'Periodic trig 1us'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_34/', 'title': 'Periodic cont 1us'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_11/', 'title': 'Periodic trig 5us'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_23/', 'title': 'Periodic cont 5us'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_15/', 'title': 'Periodic trig 10us'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_27/', 'title': 'Periodic cont 10us'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_19/', 'title': 'Periodic trig 20us'},
#     {'event_rate': 5000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_31/', 'title': 'Periodic cont 20us'}
# )

# 2 MHz pp
# sim_runs = (
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_2/',  'title': 'MB trig 100ns'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_36/', 'title': 'Periodic trig 1us'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_33/', 'title': 'Periodic cont 1us'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_10/', 'title': 'Periodic trig 5us'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_22/', 'title': 'Periodic cont 5us'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_14/', 'title': 'Periodic trig 10us'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_26/', 'title': 'Periodic cont 10us'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_18/', 'title': 'Periodic trig 20us'},
#     {'event_rate': 2000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_30/', 'title': 'Periodic cont 20us'}
# )

# 1 MHz pp
#sim_runs = (
#    {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_1/',  'title': 'MB trig 100ns'},
#    {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_35/', 'title': 'Periodic trig 1us'},
#    {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_32/', 'title': 'Periodic cont 1us'},
#    {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_9/',  'title': 'Periodic trig 5us'},
#    {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_21/', 'title': 'Periodic cont 5us'},
#    {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_13/', 'title': 'Periodic trig 10us'},
#    {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_25/', 'title': 'Periodic cont 10us'},
#    {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_17/', 'title': 'Periodic trig 20us'},
#    {'event_rate': 1000,  'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_29/', 'title': 'Periodic cont 20us'}
#)

# 400 kHz pp
# sim_runs = (
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_0/',  'title': 'MB trig 100ns'},    
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_8/',  'title': 'Periodic trig 5us'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_20/', 'title': 'Periodic cont 5us'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_12/', 'title': 'Periodic trig 10us'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_24/', 'title': 'Periodic cont 10us'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_16/', 'title': 'Periodic trig 20us'},
#     {'event_rate': 400,   'path': '/scratch/ITS_SystemC/sim_output_its_pp/run_28/', 'title': 'Periodic cont 20us'}
# )

            
#sim_type = "pp"
#sim_additional_title = "400 kHz pp"
#plot_output_dir = '/scratch/ITS_SystemC/plots_pp/data_rate_plots/'



save_plots = True

# Create plot output directory
if save_plots:
    if not os.path.exists(plot_output_dir):
        os.makedirs(plot_output_dir)

# Dict of dict:
# Outer most dict: key: event_rate, value: dict
# Inner most dict: key: layer, value: data rate
sim_data_mean = dict() # per link
sim_data_stddev = dict() # per link
sim_protocol_mean = dict() # per link
sim_protocol_stddev = dict() # per link
sim_total_data = dict() # total for stave/RU
sim_total_protocol = dict() # total for stave/RU

data_links_per_stave_per_layer = {0: 9,
                                  1: 9,
                                  2: 9,
                                  3: 16,
                                  4: 16,
                                  5: 28,
                                  6: 28}

for sim_num in range(len(sim_runs)):
    # Check config, see how many staves per layer were simulated
    config = configparser.ConfigParser()
    config.read(sim_runs[sim_num]['path'] + '/settings.txt')
    
    num_events = config.getint('simulation','n_events')
    avg_event_rate_ns = config.getint('event','average_event_rate_ns')
    sim_time_ns = num_events*avg_event_rate_ns
    sim_time_seconds = sim_time_ns/1.0E9
    
    print('sim time: ' + str(sim_time_seconds))
    
    dict_data_mean = dict()
    dict_data_stddev = dict()
    dict_protocol_mean = dict()
    dict_protocol_stddev = dict()
    dict_total_data = dict()
    dict_total_protocol = dict()
    
    for layer in range(7):
        num_staves = config.getint('its','layer' + str(layer) + '_num_staves')    
    
        if num_staves == 0:
            dict_data_mean[layer] = 0
            dict_data_stddev[layer] = 0
            dict_protocol_mean[layer] = 0
            dict_protocol_stddev[layer] = 0
            dict_total_data[layer] = 0
            dict_total_protocol[layer] = 0
        else:
            links_per_stave = data_links_per_stave_per_layer[layer]
            
            data_per_link_arr = np.zeros(links_per_stave*num_staves)
            protocol_per_link_arr = np.zeros(links_per_stave*num_staves)
            
            for stave in range(num_staves):
                # Open link utilization file for current layer/stave
                with open(sim_runs[sim_num]['path'] + '/RU_' + str(layer) + '_' + str(stave) + '_Link_utilization.csv') as csv_file:
                    csv_reader = csv.DictReader(csv_file, delimiter=';')
                    link_num = 0
                    for row in csv_reader:
                        data_bytes = int(row['DATA_SHORT (bytes)']) + \
                                     int(row['DATA_LONG (bytes)']) + \
                                     int(row['REGION_HEADER (bytes)'])
                        
                        protocol_bytes = int(row['CHIP_HEADER (bytes)']) + \
                                         int(row['CHIP_TRAILER (bytes)']) + \
                                         int(row['CHIP_EMPTY_FRAME (bytes)']) + \
                                         int(row['BUSY_ON (bytes)']) + \
                                         int(row['BUSY_OFF (bytes)'])
                        
                        #print('Layer: ' + str(layer) + ' stave: ' + str(stave) + ' link: ' + str(link_num))
                        
                        data_mbps = (float(data_bytes) * 8) / (1.0E6 * sim_time_seconds)
                        protocol_mbps = (float(protocol_bytes) * 8) / (1.0E6 * sim_time_seconds)
                        
                        #data_mbps = (float(data_bytes) * 8) / ((1024*1024) * sim_time_seconds)
                        #protocol_mbps = (float(protocol_bytes) * 8) / ((1024*1024) * sim_time_seconds)
                        
                        data_per_link_arr[stave*links_per_stave + link_num] = data_mbps
                        protocol_per_link_arr[stave*links_per_stave + link_num] = protocol_mbps
                        
                        link_num += 1
            
            # Data per link per layer
            dict_data_mean[layer] = np.mean(data_per_link_arr)                        
            dict_data_stddev[layer] = np.std(data_per_link_arr)
            dict_protocol_mean[layer] = np.mean(protocol_per_link_arr)                        
            dict_protocol_stddev[layer] = np.std(protocol_per_link_arr)
            
            # Avg data per stave/RU per layer
            # Maybe I should calculate stddev here as well..
            dict_total_data[layer] = np.sum(data_per_link_arr)/num_staves
            dict_total_protocol[layer] = np.sum(protocol_per_link_arr)/num_staves
    
    sim_data_mean[sim_num] = dict_data_mean
    sim_data_stddev[sim_num] = dict_data_stddev
    sim_protocol_mean[sim_num] = dict_protocol_mean
    sim_protocol_stddev[sim_num] = dict_protocol_stddev
    sim_total_data[sim_num] = dict_total_data
    sim_total_protocol[sim_num] = dict_total_protocol
    

# Plot data rate per link

fig = plt.figure(figsize=(20,10))
plt.style.use('seaborn-whitegrid')

width = (1.0/len(sim_runs)) * 0.8

plots_data = dict()
plots_protocol = dict()


# Plot with separate error bars
plot_num = 0
for sim_run in sim_data_mean:
    list_data_mean = sorted(sim_data_mean[sim_run].items())
    list_data_stddev = sorted(sim_data_stddev[sim_run].items())
    list_protocol_mean = sorted(sim_protocol_mean[sim_run].items())
    list_protocol_stddev = sorted(sim_protocol_stddev[sim_run].items())

    layers, data_mean = zip(*list_data_mean)
    layers, data_stddev = zip(*list_data_stddev)
    layers, protocol_mean = zip(*list_protocol_mean)
    layers, protocol_stddev = zip(*list_protocol_stddev)
    
    print(type(layers))
    print(layers)
    
    layers_arr = np.array(layers)
    
    plots_data[sim_run] = plt.bar(layers_arr + (plot_num-len(sim_runs)/2)*width, data_mean, width,
                                  yerr=data_stddev, edgecolor='black', linewidth=1, color='C'+str(sim_run), 
                                  label=sim_runs[sim_run]['title'])
    
    plots_protocol[sim_run] = plt.bar(layers_arr + (plot_num-len(sim_runs)/2)*width, protocol_mean, width,
                                      bottom=data_mean, yerr=protocol_stddev, edgecolor='black', color='C'+str(sim_run),
                                      hatch='///', linewidth=1, linestyle='dashed', label=None)

    plot_num += 1


additional_title = ''
if len(sim_additional_title) > 0:
    additional_title = ' - ' + sim_additional_title

plt.title('Data rate per link' + additional_title, fontsize=45)
plt.xlabel('Layer', fontsize=35)
plt.ylabel('Data rate [Mbps]', fontsize=35)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc='upper right', fontsize=25)
#plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=20)

#ax.yaxis.set_major_locator(MultipleLocator(100))
#ax.yaxis.grid(True, which='minor')
#ax.yaxis.set_minor_locator(MultipleLocator(50))

plt.tight_layout()
plt.show()

if save_plots:
    fig.savefig(plot_output_dir + '/data_rates_per_link_sep_error.png', bbox_inches='tight', facecolor=fig.get_facecolor(), edgecolor='none')
    fig.savefig(plot_output_dir + '/data_rates_per_link_sep_error.pdf', bbox_inches='tight', facecolor=fig.get_facecolor(), edgecolor='none')
plt.close(fig)


# Plot with shared error bars
fig = plt.figure(figsize=(20,10))
plt.style.use('seaborn-whitegrid')

width = (1.0/len(sim_runs)) * 0.8

plots_data = dict()
plots_protocol = dict()

plot_num = 0
for sim_run in sim_data_mean:
    list_data_mean = sorted(sim_data_mean[sim_run].items())
    list_data_stddev = sorted(sim_data_stddev[sim_run].items())
    list_protocol_mean = sorted(sim_protocol_mean[sim_run].items())
    list_protocol_stddev = sorted(sim_protocol_stddev[sim_run].items())

    layers, data_mean = zip(*list_data_mean)
    layers, data_stddev = zip(*list_data_stddev)
    layers, protocol_mean = zip(*list_protocol_mean)
    layers, protocol_stddev = zip(*list_protocol_stddev)
    
    shared_stddev = [sum(x) for x in zip(data_stddev, protocol_stddev)]
    
    print(type(layers))
    print(layers)
    
    layers_arr = np.array(layers)
    
    plots_data[sim_run] = plt.bar(layers_arr + (plot_num-len(sim_runs)/2)*width, data_mean, width,
                                  edgecolor='black', linewidth=1, color='C'+str(sim_run), 
                                  label=sim_runs[sim_run]['title'])
    
    plots_protocol[sim_run] = plt.bar(layers_arr + (plot_num-len(sim_runs)/2)*width, protocol_mean, width,
                                      bottom=data_mean, yerr=shared_stddev, edgecolor='black', color='C'+str(sim_run),
                                      hatch='///', linewidth=1, linestyle='dashed', label=None)

    plot_num += 1


additional_title = ''
if len(sim_additional_title) > 0:
    additional_title = ' - ' + sim_additional_title

plt.title('Data rate per link' + additional_title, fontsize=45)
plt.xlabel('Layer', fontsize=35)
plt.ylabel('Data rate [Mbps]', fontsize=35)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc='upper right', fontsize=25)
#plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=20)

#ax.yaxis.set_major_locator(MultipleLocator(100))
#ax.yaxis.grid(True, which='minor')
#ax.yaxis.set_minor_locator(MultipleLocator(50))

plt.tight_layout()
plt.show()

if save_plots:
    fig.savefig(plot_output_dir + '/data_rates_per_link_shared_error.png', bbox_inches='tight', facecolor=fig.get_facecolor(), edgecolor='none')
    fig.savefig(plot_output_dir + '/data_rates_per_link_shared_error.pdf', bbox_inches='tight', facecolor=fig.get_facecolor(), edgecolor='none')
plt.close(fig)





# Plot data rate per stave/RU
fig = plt.figure(figsize=(20,10))
plt.style.use('seaborn-whitegrid')

width = (1.0/len(sim_runs)) * 0.8

plots_data = dict()
plots_protocol = dict()

plots_total_data = dict()
plots_total_protocol = dict()
plot_num = 0

for sim_run in sim_data_mean:
    list_total_data = sorted(sim_total_data[sim_run].items())
    list_total_protocol = sorted(sim_total_protocol[sim_run].items())

    layers, data_total = zip(*list_total_data)
    layers, protocol_total = zip(*list_total_protocol)
    
    print(type(layers))
    print(layers)
    
    layers_arr = np.array(layers)

    plots_total_data[sim_run] = plt.bar(layers_arr + (plot_num-len(sim_runs)/2)*width, data_total, width,
                                        edgecolor='black', linewidth=1, color='C'+str(sim_run),
                                        label=sim_runs[sim_run]['title'])
    
    plots_total_protocol[sim_run] = plt.bar(layers_arr + (plot_num-len(sim_runs)/2)*width, protocol_total, width,
                                            edgecolor='black', linewidth=1, linestyle='dashed', color='C'+str(sim_run),
                                            hatch='///', bottom=data_total, label=None)

    plot_num += 1
    
plt.title('Data rate per stave/RU' + additional_title, fontsize=45)
plt.xlabel('Layer', fontsize=35)
plt.ylabel('Data rate [Mbps]', fontsize=35)
plt.legend(loc='upper right', fontsize=25)
#plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=25)

ax = fig.gca()

ax.yaxis.grid(True, which='major', linewidth=2)
ax.yaxis.set_major_locator(MultipleLocator(3200))
ax.yaxis.set_major_formatter(NullFormatter())
ax.yaxis.grid(True, which='minor')
ax.yaxis.set_minor_locator(MultipleLocator(800))
#ax.yaxis.set_minor_locator(MultipleLocator(200))
ax.yaxis.set_minor_formatter(FormatStrFormatter('%d'))
ax.tick_params(axis='both', which='major', labelsize=25)
ax.tick_params(axis='both', which='minor', labelsize=25)

plt.tight_layout()
plt.show()

if save_plots:
    fig.savefig(plot_output_dir + '/data_rates_per_stave.png', bbox_inches='tight', facecolor=fig.get_facecolor(), edgecolor='none')
    fig.savefig(plot_output_dir + '/data_rates_per_stave.pdf', bbox_inches='tight', facecolor=fig.get_facecolor(), edgecolor='none')
plt.close(fig)
```
